## Chapters 7-8: From GPS to Surveillance

Pages 143-185 of *Superagency* shift from abstract innovation principles to concrete technologies: GPS as informational infrastructure and code as architectural control. Hoffman and Beato argue that AI functions like GPS for language - helping navigate from point A to point B with greater certainty.

The chapters made me uncomfortable in a different way than earlier ones. Not because they're wrong, but because they're *persuasive*. I'm starting to see the optimistic case even as I remain skeptical about who benefits.


## Chapter 7: GPS as Democratic Infrastructure

Hoffman highlights that GPS generated $1.4 trillion in economic benefits from 1984 to 2017. The technology emerged from military origins but became publicly accessible infrastructure.

The FCC requirement in the early 2000s - that cell providers must give 911 call centers GPS coordinates - accelerated phone-based GPS adoption. Policy drove technology adoption, which created new possibilities.

This is the optimistic narrative: regulation can steer innovation toward public benefit.

**My question:** If GPS created $1.4 trillion in value, who captured most of it? The economic benefit number says nothing about distribution.


## The Ability to Get Lost

Hoffman asks: "Do we reduce the scope of human existence by making the world so legible we completely lose our ability to get lost?"

This resonated with me more than it should have.

Anti-AI rhetoric often centers on loss of humanness - our ability to create, feel, explore, be sentient. It comes from fear, and rightfully so. We've watched jobs disappear for the benefit of the few.

But the book is making me ask: what about the flipside? What about the benefits that aren't sellable news stories? History focuses on dramatic failures. Media amplifies fear because it grabs attention.

Is that the right lens?

I don't know yet. But I'm less certain than I was.


## AI as GPS for Language

The book positions AI as a system for "analyzing, synthesizing, and mapping language flows" - helping navigate from point A to point B with greater efficiency.

The analogy has limits.

GPS coordinates are objective. Language is not. The "informational planet" of each LLM differs based on training data, algorithms, weights, and architectural choices. There's no shared ground truth.

GPS works because we agree on latitude and longitude. AI outputs diverge because we don't agree on what language *means*.

The book acknowledges this but doesn't dwell on it long enough.


## Chapter 8: Code is Law

Hoffman traces the internet's origins as open-source and decentralized. Commercialization changed that. User authentication, identity verification, centralization - all became necessary for commerce.

Lawrence Lessig's framework: four constraints regulate behavior - laws, markets, norms, and *architecture*. On the internet, architecture plays an outsized role. **"Code is law."**

Hoffman's concern: AI broadens the possibilities of perfect control.

Self-driving cars that won't exceed speed limits. Smart contracts that execute without negotiation. Surveillance systems that track compliance continuously.

**The question:** Is there liberty that emerges from inefficiencies in enforcement?


## The Uncontract

Hoffman borrows from Shoshana Zuboff's *The Age of Surveillance Capitalism* to introduce the "uncontract" - computer-mediated agreements where code replaces human judgment.

Zuboff: "The uncontract desocialized the contract, manufacturing certainty through the substitution of automated procedures for promise, dialogue, shared meaning, problem solving, dispute resolution, and trust."

Self-executing agreements make judgment, negotiation, flexibility, moral reasoning, forgiveness, and empathy superfluous.

This is the slippery slope the book is warning about. Not dramatic AI takeover. Gradual erosion of human discretion.


## Perfect Control vs. Human Agency

The book argues that ubiquitous control has implications for privacy, autonomy, and liberty.

Society depends on collective agreement to norms - speed limits, no smoking zones, drunk driving laws. Policing can't verify every minute of our lives, so compliance relies on internalized norms and imperfect enforcement.

**What happens when enforcement becomes perfect?**

Hoffman doesn't say we shouldn't have laws. He asks whether continuous, automated compliance removes something essential about human choice - including the choice to break a law.

I'm torn on this. Perfect enforcement sounds appealing when applied to drunk driving. It sounds dystopian when applied to everything else.

Where's the line?


## Can We Be Optimistic?

The book wants me to believe that AI - like GPS - can be democratizing. That it provides access to services previously reserved for the highest classes. That regulation can guide it toward public benefit.

I'm starting to see the argument. But I'm not convinced distribution follows innovation automatically.

GPS created $1.4 trillion in value. How much went to military contractors? Tech companies? Wealthier users with smartphones?

The book focuses on aggregate benefit. I keep asking: **who benefits?**

Hoffman argues we should develop AI "in alignment with democratic ideals of self-determination and broad participation."

Sure. But how? That's the question the book keeps deferring.


## Where I'm At

These chapters are persuasive because they're *grounded*. GPS is real infrastructure with measurable benefits. Code-as-architecture is observable in every platform we use.

The optimistic case is stronger than I expected. But the distributional questions remain unanswered.

Hoffman sees GPS and thinks: AI can be like this - regulated, publicly beneficial, democratizing.

I see GPS and think: Who owned the satellites? Who profited from the FCC mandate? Who got smartphones first?

The book is making me less reflexively skeptical. But it hasn't convinced me that innovation distributes benefit fairly without intentional policy forcing it.

**Next chapters will determine whether optimism is justified or just well-argued.**
