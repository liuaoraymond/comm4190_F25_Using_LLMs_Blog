## The Promise: AI-Generated Study Podcasts

For my Leading Across Differences class, we had 20+ academic readings. I tried Google's NotebookLM, which generates audio podcast summaries of uploaded documents.

The idea: Listen to summaries instead of reading everything.

It didn't work.

## What I Tried

I uploaded all 15 readings from one week - a mix of academic articles, case studies, and book chapters on leadership, identity, and power dynamics.

NotebookLM generated a 10-minute "podcast" summarizing the key themes across all readings.

Two AI voices had a conversational discussion about the material. It sounded professional and engaging.


## What Went Wrong

### It Missed Crucial Details
The class required three essays analyzing specific arguments from specific authors. NotebookLM's summary mentioned themes but didn't preserve:
- Who said what
- Which study supported which claim
- Distinctions between competing frameworks

I knew the general concepts but couldn't cite anyone.

### 10 Minutes Can't Cover 15 Readings
Each reading was 20-40 pages. That's 300+ pages of dense academic writing.

NotebookLM condensed it into 10 minutes, which means it selected what it thought was important and dropped the rest.

For a multiple-choice test, maybe that works. For essays where I need to synthesize arguments, it didn't.

### The Format Was Too Passive
Listening to a podcast is easy. Too easy. I retained almost nothing.

Reading forces active engagement. I highlight, annotate, pause to think. Audio just washes over you.


## When NotebookLM Might Actually Help

**Good for:**
- Refreshing material you've already read
- Getting a high-level overview before diving deep
- Supplementing notes (not replacing reading)
- Subjects with clear factual content (not interpretive analysis)

**Not good for:**
- Classes requiring detailed citations
- Essay-based assessments
- Dense theoretical reading
- First-pass learning of complex material


## What I Use Instead

**For readings:** I actually read them. I highlight key arguments and take notes in margins.

**For review:** I make my own summaries after reading - writing forces comprehension.

**For exam prep:** I create concept maps connecting ideas across readings. NotebookLM can't do this because it doesn't understand *my* interpretation.

**Audio tools:** I use text-to-speech to listen while I follow along in the document. That's different from listening to AI-generated summaries.


## The Efficiency Trap

NotebookLM saved time. But it didn't save learning.

I spent 10 minutes listening to a summary and understood nothing deeply. My classmates who read spent 3 hours and could actually write the essays.

**The extra time mattered.**

Maybe NotebookLM works better for straightforward material - timelines, definitions, factual summaries. But for interpretive work where you need to form your own arguments? AI summaries don't replace the cognitive work of reading.

Efficiency isn't always effective.
