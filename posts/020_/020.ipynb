## A Study That Made Me Uncomfortable
I came across MIT Media Lab research on how ChatGPT affects brain activity during essay writing, and it forced me to reflect on my own AI usage.
The study tracked 54 participants over four months as they wrote essays using LLMs, search engines, or no tools at all. EEG data showed that LLM users consistently had the weakest neural connectivity.


## What the Study Found

### Brain Connectivity Differences
- **Brain-only:** strongest, most distributed activation
- **Search Engine:** moderate engagement
- **LLM-assisted:** lowest engagement, minimal connectivity

The more a tool did the thinking, the less the brain engaged.

### Cognitive Debt
When participants switched tools:
- LLM-to-Brain users struggled to activate key networks
- Brain-to-LLM users still showed higher engagement than chronic LLM users

The researchers called this pattern **cognitive debt**: a downward shift in mental effort that builds up with repeated offloading.

## The Ownership Problem
Participants rated how much the work felt like their own:
- Brain-only: highest ownership
- Search Engine: moderate
- LLM users: lowest

LLM users also failed to accurately quote their own essays later. They had not synthesized the material themselves, so the memory trace was weaker.


## Everyone Started Sounding the Same
Using natural language processing, the study found that essays written with LLM assistance were more similar to each other:
- similar topic structures
- similar n-gram patterns
- similar named entities

LLM-assisted writing flattened personal style.

## Why This Makes Me Uncomfortable
I have relied heavily on Claude for structuring notes, drafting responses, and shaping arguments.
This study suggests that I may have been outsourcing the very cognitive work that builds skill, memory, and identity as a writer.

When I revise an AI-generated draft, I am in approval mode, not creation mode. My prefrontal cortex does less work.


## Learning Implications
### Cognitive Debt Accumulates
Long-term LLM reliance leads to:
- lower neural activation
- weaker memory formation
- reduced cognitive engagement
- decreased ownership

The brain adapts to the tools it uses.


## The Larger Picture
### Who Benefits From Cognitive Offloading?
Productivity increases on paper. But learning decreases.
In education, cognitive effort *is* the point.
LLMs replace the work that builds thinking capacity.

### The Convenience Trap
You do not feel cognitive decline. You only feel the difficulty when the tool is removed.
This happened in the experiment: LLM-to-Brain users struggled without assistance.


## What I'm Thinking About Now
### Where I Will Reduce AI
- tasks requiring learning
- tasks requiring synthesis
- tasks requiring creativity

### Where AI Is Fine
- scheduling
- logistics
- routine organization

The goal is to use AI without letting it replace my thinking.


## Final Thoughts
The MIT study raises serious concerns about what we lose when AI handles the thinking.
We gain speed and convenience.
But we risk losing:
- cognitive engagement
- memory formation
- ownership
- original thought

**The question is not whether AI works. It is what happens to our minds when we let it think for us.**
