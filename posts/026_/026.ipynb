## The Problem: Planning a 11-Day Japan Trip

I'm traveling to Japan December 13-23. I wanted an itinerary focused on architecture, design, and niche culinary experiences. I tried ChatGPT first.

It didn't work.

## What I Asked ChatGPT

"Plan a 11-day Japan itinerary (Dec 13-23) focused on:
- Modern architecture and design
- Niche culinary spots (not tourist traps)
- Efficient train routes between cities"

ChatGPT generated a detailed day-by-day plan with recommendations, transit times, and restaurant suggestions.

On paper, it looked perfect.


## What Went Wrong

### Train Dates Were Wrong
ChatGPT suggested specific trains and departure times. When I checked the actual JR schedules, half the trains didn't run on those dates or times.

I would've missed connections if I'd trusted the itinerary blindly.

### It Missed My Vibe Completely
**What I wanted:** Tadao Ando buildings, small design shops, local izakayas recommended by residents

**What ChatGPT gave me:** Tokyo Skytree, Senso-ji Temple, Tsukiji Fish Market - all mainstream tourist spots

The recommendations weren't *wrong*. They just weren't *mine*.

### The Culinary Suggestions Were Generic
ChatGPT recommended well-known ramen chains and Michelin-starred restaurants. Nothing niche. Nothing local.

When I asked for "hidden gem food spots," it gave me places with 10,000+ Google reviews.

### The Outputs Were Overwhelming
Each day had 8-10 activities listed with paragraph descriptions. Reading through it felt like homework.

I couldn't see the flow of the trip. Just a wall of text with too many options.


## What Actually Worked

I ended up using:

**TikTok and YouTube:** Locals and frequent travelers showing actual spots. I found a tiny coffee shop in Kyoto that specializes in architecture books, a hidden kaiseki restaurant in Osaka, and a design district ChatGPT never mentioned.

**Google Maps + Word of Mouth:** I searched for "architecture" near areas I was already visiting. User reviews from real people who'd been there recently.

**Travel blogs by residents:** Japanese design blogs and expat food guides gave me the niche spots AI couldn't.

**Manual train checking:** I verified every train route on the actual JR website. Time-consuming but necessary.


## Why AI Itinerary Planning Failed

### It Optimizes for Popularity, Not Preference
ChatGPT recommends places that appear frequently in its training data - which means tourist-heavy spots dominate.

Niche recommendations require local knowledge AI doesn't have.

### It Can't Verify Real-Time Information
Train schedules change. Restaurants close. Events get canceled. AI doesn't know this.

It generates confident answers based on outdated data.

### It Doesn't Understand "Vibe"
I can tell ChatGPT I like "architecture and design," but it can't distinguish between:
- A famous building everyone photographs
- A quiet neighborhood with innovative residential projects

That nuance requires human curation.


## When AI Itinerary Tools Might Work

**Good for:**
- High-level structure (how many days in each city)
- General logistics (visa requirements, currency, weather)
- First-time travelers who want mainstream highlights

**Not good for:**
- Niche interests
- Real-time transit schedules
- Finding hidden local spots
- Balancing pacing and downtime

I spent 3 hours with ChatGPT and got an unusable itinerary. I spent 5 hours manually researching and got a trip I'm actually excited about.

**The extra time mattered.** AI gave me efficiency. Manual research gave me a trip that fits who I am.
