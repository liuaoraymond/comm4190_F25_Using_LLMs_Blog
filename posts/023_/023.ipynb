## The Assignment: Advertise a Product That Doesn't Exist

For my Marketing and Creativity class, I needed to create advertising concepts for "Bulby" - a climate-responsive smart lightbulb. The product isn't real. I needed mockups showing what the campaigns would look like.

I used Google DeepMind's Veo to generate the visuals.

## What I Needed

Two campaign concepts:

**Physical:** Times Square billboard showing "It's hot, Bulby" - using extreme heat as contrast to the product's climate awareness

**Digital:** Social media ad showing "It's cold, Bulby" - clean, modern living room with the lightbulb responding to temperature

Photoshop would take hours. Hiring a designer costs money I don't have. Veo generated both in 20 minutes.

## What Worked

### The Times Square Billboard
Prompt: "Large billboard in Times Square showing a glowing smart lightbulb with text 'It's hot, Bulby' - evening, city lights, realistic"

Veo nailed the environmental context. The billboard looked like it belonged there. Lighting, perspective, urban density - all correct.

### The Living Room Scene
Prompt: "Modern minimalist living room, large windows, smart lightbulb visible with temperature icon, text 'It's cold, Bulby' - bright natural lighting"

Clean, professional, Instagram-ready aesthetic. Exactly what I needed for a social media mockup.

## What Didn't Work

### Text Rendering
Veo can't do text reliably. "It's hot, Bulby" came out garbled half the time. I had to regenerate 6-8 times per image.

### Product Details
The lightbulb design kept changing. I needed consistency across campaigns but Veo generated different bulb shapes each time.

I ended up overlaying the same product image manually in post.

### Brand Consistency
Colors, fonts, styling - all inconsistent between generations. AI doesn't understand brand guidelines.

## What This Actually Replaced

**Not:** Creative direction, concept development, copywriting

**Yes:** Stock photography, staging physical mockups, hiring a designer for proof-of-concept

I already knew what I wanted the ads to look like. Veo let me visualize concepts fast enough to iterate.

For a class project where the point is the marketing strategy, not the production quality, this worked fine.

## When AI Mockups Make Sense

**Good for:**
- Proof-of-concept visuals
- Class assignments where execution isn't the focus
- Testing multiple creative directions quickly
- Environmental context (cityscapes, interiors)

**Not good for:**
- Final production assets
- Consistent product rendering
- Text and typography
- Brand guidelines adherence

I got mockups good enough to present the advertising strategy. That's all I needed.

The product concept mattered more than pixel-perfect execution.
